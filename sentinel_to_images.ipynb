#Image Processing

```python
def normalize(band):
    band_min, band_max = (band.min(), band.max())
    return ((band-band_min)/((band_max - band_min)))

def brighten(band):
    alpha=0.13
    beta=0
    return np.clip(alpha*band+beta, 0,255)

def gammacorr(band):
    gamma=2
    return np.power(band, 1/gamma)
```

```python
data_folder = '/content/drive/MyDrive/Varuna Hackathon 2022/sentinel-2 image/'
all_folders = [ data_folder + '2020/', data_folder + '2021/']
          
for sub_folder in all_folders:
  year = sub_folder[-5:-1]
  saved_folder = '/content/' + year

  all_dates = os.listdir(sub_folder)

  for date in all_dates:
    print(date)
    blue_path = os.path.join(sub_folder, date, 'IMG_DATA/47PQS_'+ date + '_B02.jp2')
    green_path = os.path.join(sub_folder, date, 'IMG_DATA/47PQS_'+ date + '_B03.jp2')
    red_path = os.path.join(sub_folder, date, 'IMG_DATA/47PQS_'+ date + '_B04.jp2')

    blue = rasterio.open(blue_path).read(1)
    green = rasterio.open(green_path).read(1)
    red = rasterio.open(red_path).read(1)

    blue_n = normalize(blue)
    green_n = normalize(green)
    red_n = normalize(red)
    
    blue_b = brighten(blue)
    green_b = brighten(green)
    red_b = brighten(red)
    
    blue_bn = normalize(blue_b)
    green_bn = normalize(green_b)
    red_bn = normalize(red_b)

    rgb_composite_n = np.dstack((red_n, green_n, blue_n))
    rgb_composite_bn = np.dstack((red_bn, green_bn, blue_bn))

    rgb_plot=plt.imshow(rgb_composite_n, interpolation='lanczos')
    plt.axis('off')
    plt.savefig(saved_folder + '/rgb/' + date + '.png', dpi=200, bbox_inches='tight')
    plt.close('all')

    rgb_plot=plt.imshow(rgb_composite_bn, interpolation='lanczos')
    plt.axis('off')
    plt.savefig(saved_folder + '/rgb_bn/' + date + '.png', dpi=200, bbox_inches='tight')
    plt.close('all')
```
